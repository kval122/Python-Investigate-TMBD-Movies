### Cleaning Data

```python
# Checking how many cells of each column are missing data

df.isna().sum()
```

```python
#Checking how many cells are unique in each column

df.nunique()
```

##### We can see here that the best column to use as a primary key is the id column. There is only one cell that is duplicated which we will inspect below. 

```python
# View duplicated id row

df[df.duplicated(subset = 'id', keep =False)].sort_values('id')
```

```python
# These row do appear to be identical, so we will drop row with index number 2090
df = df.drop(df.index[2089])
```

***

##### Because we have determined to use the id as a primary key, we can drop the imdb_id and the original_title

```python
# Drop columns imbd_id and original_title

df = df.drop(labels=['imdb_id', 'original_title'], axis=1)
df.info()
```

***

##### The columns with qualitative data missing that we will not be using for analysis are cast, genres and production companies, overview, homepage, keywords and tagline. We will drop those below

```python
# Drop not applicable qualitative columns
df = df.drop(labels=['cast', 'genres', 'production_companies', 'overview', 'homepage', 'keywords', 'tagline', 'director'], axis=1)
df.info()
```

```python
# Excellent, lets take a peek at our data now

df.head(20)
```

***

##### Because we are comparing data from many decades, the budget and revenue including inflation are the best columns to use when examining finances. We will drop 'budget' and 'revenue' and keep 'budget_adj' and 'revenue_adj'. 

```python
# Drop budget and revenue columns

df = df.drop(labels=['revenue', 'budget'], axis=1)
df.head()
```

***

##### Checking data for mistypes or 0 values

```python
# Statistical data 

df.describe()
```

##### The columns 'runtime', 'budget_adj' and 'revenue_adj' have zero values as their minimum when this is unlikely to be accurate. 

```python
# Using a vectorized approach to reduce coding, Counting the number of zero values in each column
(df==0).sum()
```

```python
# Drop revenue_adj and budget_adj

df = df.drop(labels=['revenue_adj', 'budget_adj'], axis=1)
df.info()
```

## Exploratory Data Analysis


```python
# Comparing all numerical variables for a general idea of relationship
pd.option_context('mode.use_inf_as_na', True)
sns.pairplot(df, vars=df.columns[2:]);
```

### Question 1. Does the runtime of the movie have an effect on the vote average?

```python
# Calculate average vote average per runtime and its corresponding counts
x_counts = df.groupby('runtime')['runtime'].count()
average_vote_per_runtime = df.groupby('runtime')['vote_average'].mean()

# Create a bubble chart
plt.figure(figsize=(10, 6))

# Set x-axis range
plt.xlim(0, 350)  

# Scale bubble size based on count of runtime-values
bubble_sizes = x_counts.values * 50

plt.scatter(average_vote_per_runtime.index, average_vote_per_runtime.values, s=bubble_sizes, alpha=0.7)
plt.xlabel('Runtime (minutes)')
plt.ylabel('Vote Average')
plt.title('Average Vote Average vs. Runtime')
plt.grid(True)
plt.show()
```

##### This chart shows that movies with a runtime between 200 and 250 minutes are more likely to have a higher rating
***

```python
def myGrpPlot(df, grpVar, yVar, stat='mean', kind='line', figsize=[14,6]):
    '''
    """
    Creates a plot for a grouped variable against another variable in a pandas DataFrame.

    Args:
        df (pd.DataFrame): The pandas DataFrame containing the data to be plotted.
        grpVar (str): The name of the column in the DataFrame to group by.
        yVar (str): The name of the column in the DataFrame to aggregate and plot on the y-axis.
        stat (str, optional): The aggregation statistic to apply to the y variable. Defaults to 'mean'.
            Can be any valid aggregation function supported by pandas.GroupBy.agg().
        kind (str, optional): The type of plot to create. Defaults to 'line'. 
            Can be 'line' or 'bar'.
        figsize (list[float], optional): A list of two elements representing the width and height of the figure 
            in inches. Defaults to [14, 6].

    Returns:
        None

    Creates a plot using Matplotlib based on the specified group variable, target variable, 
    aggregation statistic, plot type, and figure size. It automatically formats labels by 
    replacing underscores with spaces, adds a title, and formats x and y labels with bold font.
    '''

    #plot options
    kwargs={'edgecolor':'black','rot':0,'width':0.08} if kind=='bar' else{}
    #plot
    df.groupby(grpVar)[yVar].agg(stat).plot(kind=kind, figsize=figsize, **kwargs);
    #Tidy up labels by replacing _ with a space
    grpVar=grpVar.replace("_"," ")
    yVar=yVar.replace("_"," ")
    #Add Title and format it
    plt.title(f'{stat} of {yVar} by {grpVar}'.title(), fontsize= 14, weight = "bold")
    #Add X and Y labels with formatting
    plt.xlabel(grpVar.title(), fontsize= 10, weight = "bold")
    plt.ylabel(f'{stat} {yVar}'.title(), fontsize = 10, weight = "bold")
    
```

### Question 2. How has runtime changed over the years?

```python
myGrpPlot(df,'release_year', 'runtime')
```

##### This chart very clearly shows that the average runtime of movies is decreasing overtime
***

### Question 3. How has vote average changed over the years?

```python
myGrpPlot(df,'release_year', 'vote_average')
```

#### As seen in this chart, the average vote has decreased over the years.
